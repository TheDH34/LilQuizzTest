```python
def question(question, reponse, nb_chances_restantes):
    """
    Pose une question à l'utilisateur et gère la vérification et chances restantes.

    question: la question à afficher.
    reponse: La réponse correcte (doit être une chaîne de caractères).
    nb_chances_restantes: Le nombre total de chances restantes pour le quiz.
    :return: Le nombre de chances restantes après avoir répondu à la question.
    """
    if nb_chances_restantes <= 0:
        return 0

    reponse_correcte = False #Ici on créer la fonction reponse_correct = false pour la remplacer par = True plus tard
    
    while not reponse_correcte and nb_chances_restantes > 0: #on set up nos conditions, ici tant qu'on a des chances et pas la bonne reponse on replay la question
        reponse_utilisateur = input(f"{question}") #input permet de créer la question via la fonction f-string (f"question")
        if reponse_utilisateur == reponse: #si reponse correcte
            print("Correct ! Passons a la prochaine question.")
            reponse_correcte = True #c'est ici qu'on remplace notre condition False par True nous permettant de passer la prochaine question
        else:
            nb_chances_restantes -= 1 #permet de reduire de 1 le nombre de chance a chaque fail
            if nb_chances_restantes > 0:
                print(f"Wrong ! Il te reste {nb_chances_restantes} chance(s).") 
            else:
                print("Oh non ! Tu as perdu le jeu!") #permet grace au return en dessous de ne pas avoir besoin de la fonction break

    return nb_chances_restantes #return toujours au nombre de chance c'est ce qui définis si on continu ou arrête le quizz 


def quizz():
    """
    Exécute le quiz.
    """
    nb_de_chances = 3
    print("Voici notre quiz ! Tu as trois chances au total pour toutes les questions.") 
    #ici on créer notre def quizz pour lancé le quizz on commence avec 3 chances
    nb_de_chances = question(
        "Combien de fois la France a gagné la coupe du monde ? ", #on met la question & la reponse attention la virgule pour séparer les 2
        "2", 
        nb_de_chances
    )
    if nb_de_chances <= 0:
        return #on précise a chaque fois que si le nombre de chance tombe a 0 on recommence tout

    nb_de_chances = question(
        "Quand a été fondé Apple ? ", 
        "1976", 
        nb_de_chances
    )

    if nb_de_chances <= 0:
        return

    
    nb_de_chances = question(
        "Qui a fondé SpaceX ? (Majuscule au début pour le nom & prénom) ", #précise majuscule mais on aurais pu use un .lower qui permet de transformer
        #chaque réponse en miniscule 
        "Elon Musk", 
        nb_de_chances,
    )
    
    print("Résultat Final:")
    if nb_de_chances == 3:
        print("Bravo ! Tu as gagné le quiz avec toute tes chances restantes!")
    elif nb_de_chances == 2:
        print("Bravo ! Tu as gagné le quizz avec 2 chances restantes")
    elif nb_de_chances == 1:
        print("Bravo ! Tu as gagné le quizz avec 1 chances restantes") #petit amusement de ma part pour préciser au joueur le nombre de chance restante

quizz()
```
